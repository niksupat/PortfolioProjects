```python
# Importing data
import pandas as pd

data = pd.read_csv('/Users/niksupatravanij/Desktop/Data Projects and Portfolio 2024/Insurance Fraud/Dataset.csv')
data
```

```python
# EDA - dropping null value rows
data.dropna()
```

```python
# Here we see the dataset is slightly imbalanced, with 247/909 = 20% of the data being 'Y' and the rest are 'N'
data[data['fraud_reported'] == 'Y'].shape[0]
```

```python
# Encoding data to handle non-numerical data that we want to include in the model
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
data['fraud_reported'] = label_encoder.fit_transform(data['fraud_reported'])
data['insured_sex'] = label_encoder.fit_transform(data['insured_sex'])
data['incident_type'] = label_encoder.fit_transform(data['incident_type'])
data
```

```python
# Getting variables

X = data[['insured_sex', 'incident_type', 'policy_annual_premium']]
y = data['fraud_reported']
```

```python
# Splitting the train and test data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Scaling data to improve processing in the model
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
# Fit the data to the model
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

```python
# Evaluate the model
model.score(X_test_scaled, y_test)
```

```python
y_pred = model.predict(X_test_scaled)
y_pred.shape[0]
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix

pd.DataFrame(confusion_matrix(y_test, y_pred))
```

Initially it seemed that the model was reasonably accurate at ~75%, however looking at the confusion matrix we see that it did not predict any cases of fraud. So the 75% accuracy comes from the fact that ~75% of the model is 'not fraud', so the model technically predicted these, even though it predicted all cases were not fraud.

We can try to improve this with a more complex ensemble method like the Random Forest Classifier.

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

model2 = RandomForestClassifier()
model2.fit(X_train_scaled, y_train)
```

```python
model2.score(X_test_scaled, y_test)
```

```python
y_pred2 = model2.predict(X_test_scaled)
```

```python
confusion_matrix(y_test, y_pred2)
```

Initially it looks like the RF model is less accurate than the Logistic Regression, however when we view the Confusion Matrix, we are able to see it was able to accurate predict the True Positives 11 times, which is certainly more than the 0 times in the Logistic Regression model. This could imply this is the right direction to go in, and future refinements to improve this model could be undertaken in the future.
